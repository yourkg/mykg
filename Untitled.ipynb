```python
import cv2
from PIL import Image
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
capture = cv2.VideoCapture(0)
capture.open(0)
while True:
    _,frame = capture.read()
    img = Image.fromarray(cv2.cvtColor(frame, cv2.COLOR_BGRA2RGBA))
    plt.imshow(img)
    plt.show()
    plt.cla()
capture.release()
```
